```python
def combineQS_folder(path):
    """
    Combines all query_summary files in folder path
    Returns combined qs file
    """
    
    original_path = os.getcwd()
    df = pd.DataFrame()
    
    os.chdir(path)
    for f in os.listdir(path):
        if f.find('query_summary') != -1:
            data = pd.read_csv(f)
            data = fix_df(data)
            df = pd.concat([df, data])
    df = df.reset_index(drop=True)
    os.chdir(original_path)
    return df
```

```python
# Combine all .csvs of all games
def combine_csvs(path):
    """
    Combines all .csv files that contain review data
    """
    
    original_path = os.getcwd()
    df = pd.DataFrame()
    
    os.chdir(path)
    for f in os.listdir(path):
        if f.find('query_summary') == -1:
            data = pd.read_csv(f)
            data = data.reset_index(drop=True)
            df = pd.concat([df, data], axis=0)
    df = df.reset_index(drop=True)
    os.chdir(original_path)
    return df
```
